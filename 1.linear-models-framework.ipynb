# The Principles of Statistical Modelling
Before we start discussing simple and multiple regression, it is worth taking some time to understand what we are trying to achieve at a much higher level. Both *simple* and *multiple* regression are examples of *statistical models*. Although your prior experiences may have left the impression that statistics is about *hypothesis tests*, the reality is that statistics is a science concerned with building *models* of data. This concept of model building is *fundamental* to a much deeper understanding of statistics and is fundamental to your success on the rest of this course. As such, we will spend some time here exploring what we mean by a statistical model before we go any further.

## Understanding Variance
The starting point for understanding statistical modelling is to understand the purpose of a model in terms of capturing *variance*. Variance is the basic currency of statistics. It is the fundamental information that data holds, so we need to make sure everyone has a solid handle on what *exactly* variance is.

### Defining Variance
Informally, the idea of variance is very simple. For a given set of data, we want to understand the reason why the measured values are different from each other. In other words, why does the data *vary*? If data did not vary, it would be constant. Every time we took a measurement we would get the same value and would have nothing else to say. The data would be *perfectly predictable* and there would be no need for statistics. The fact that data *does* vary is therefore fundamental for statistics.

As an example, consider the plot of all the measurements of `mpg` from the `mtcars` dataset[^foot1].

``````r
options(repr.plot.width=12, repr.plot.height=8)
``````

``````r
data('mtcars')
par(cex.lab=1.5, cex.axis=1.25, cex=1.5) # plot scaling
plot(x=seq(1,length(mtcars$mpg)), y=mtcars$mpg, xlab="Car", ylab="MPG")
``````

Clearly, the values of `mpg` differ from car-to-car. This is entirely expected, but try to consider all the reasons *why* this might be the case? 

```{admonition} What reasons can you think of for different cars having different miles per-gallon (MPG) measurements?
:class: tip, dropdown
The most obvious answer here is that cars have different properties, such as their weight or their engine size. However, there will also be a degree of natural variation due to limitations in the precision of measurement. By considering all the reasons why cars may have different values of MPG, you are starting to building your own theoretical model of the observed variance. All statistical models do is formalise this process. 
```


### Measuring Variance
In order to start understanding the variance we observe in a particular dataset, we need some way of *quantifying* it. As mentioned above, our question is really about how far any set of data are from being *constant*, given that this implies the state of *no* variance. As such, our measurement of variance needs to be in relation to some constant value. But what value to select? We could pick something arbitrary, like 0 or 100, however (for reasons we will justify further below), the better choice is to use *the mean*.

To see this visually, the mean value of `mpg` is indicated within the plot below as a horizontal line.

``````r
par(cex.lab=1.5, cex.axis=1.25, cex=1.5) # plot scaling
plot(x=seq(1,length(mtcars$mpg)), y=mtcars$mpg, xlab="Car", ylab="MPG")
abline(h=mean(mtcars$mpg))
``````

The mean is therefore our *yardstick* to measure variation against. The vertical distances from the mean to each data point tells us how much variation there is to explain. In order to get an *overall* sense of the variation across all the cars, we might first think to *sum* these distances. However, if we do that then we get a result that is essentially 0

``````r
sum(mtcars$mpg - mean(mtcars$mpg))
``````

The problem is that the mean is, by definition, *perfectly* in the middle of all the data and thus all the positve and negative distances will cancel and we will be left with nothing. As such, we instead *square* the values[^foot2] before summing them

``````r
sum((mtcars$mpg - mean(mtcars$mpg))^2)
``````

``````r
mean((mtcars$mpg - mean(mtcars$mpg))^2)
var(mtcars$mpg, use=)
``````

### The Formal Definition of Variance

Formally, the variance of the *random variable* $x$ is defined as

$$
\text{Var}\left(x\right) = E\left[\left(x - E\left[x\right]\right)^{2}\right]
$$

## Normal Probability Models

We can think of the *mean* as a very basic model of the data. It is a model that predicts a constant value for `mpg` from every car. The degree to which this is not true indicates the degree to which the data is *not* constant and thus quantifies the raw amount of variance in the data.

## Statistical Models

The fundamental elements of parametric statistical model are therefore:

- An assumed probability distribution for the data
- A mean function containing parameters that (potentially) connect the expected value of the distribution to other variables
- A variance function that captures the deviations from the expected value (the errors)

Our basic example of using the mean as a model would result in

$$
y_{i} \sim \mathcal{N}\left(\mu,\sigma\right)
$$
