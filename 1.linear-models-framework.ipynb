# The Principles of Statistical Modelling
Before we start discussing simple and multiple regression, it is worth taking some time to understand what we are trying to achieve at a much higher level. Both *simple* and *multiple* regression are examples of statistical *linear models*. In effect, every statistical analysis we run is an example of building some form of model. Although your prior experiences may have left the impression that statistics is about *hypothesis tests*, the reality is that statistics is a science concerned with building *models* of data. This concept of model building is *fundamental* to having a much deeper understanding of statistics and is fundamental to your success on the rest of this course. As such, we will spend some time here exploring what we mean by a statistical model. We will then explore simple and multiple regression as our first basic examples of linear models.

## Understanding Variance
Variance is the basic currency of statistics. It is the fundamental information that data holds, yet students sometimes do not fully appreciate what variance is and why it matters. You will hear about variance throughout this course, so we need to make sure everyone has a solid handle on what exactly it is.

### Defining Variance
Informally, the idea of variance is very straight-forward. We want to capture the reason why, for a certain set of data, the measured values are different from each other. In other words, why is data *not constant*? If data were constant, then every time we take a measurement we would get the same value and would have nothing else to say. The data would be perfectly predictable and there would be no need for statistics. As such, statistics exists *because* of the presence of variance.

As an example, consider the plot below of measurements of `mpg` from the `mtcars` dataset.

``````r
options(repr.plot.width=10, repr.plot.height=6)
``````

``````r
data('mtcars')
plot(x=seq(1,length(mtcars$mpg)), y=mtcars$mpg, xlab="Car", ylab="MPG")
``````

Clearly, the values of `mpg` differ from car-to-car. This is expected, but try to consider all the reasons *why* these values differ? Some of this variation will be due to limitations in the accuracy of measurement, however, this does not explain *all* of these variations. The obvious answer is that different cars have different properties, such as their weight or their engine size. By considering these other variables we are effectively considering different *sources* for these differences. We are doing precisely what a statistical model does: taking all the variance we can see and splitting it into different sources. The larger the chunk of variance associated with a single source, the more *importance* that source has for explaining these raw differences. 


### Measuring Variance
In order to start breaking variance apart, we need some way of *quantifying* it. As mentioned above, our question is really about how far these data are from being *constant*, given that this implies *no* variance. As such, our measurement needs to be in relation to some constant value. But what value to select? We could pick something arbitrary, like 0 or 100, however (for reasons we will justify further below), the better choice is to use *the mean*.

To see this visually, the mean value of `mpg` is indicated below as a horizontal line

``````r
plot(x=seq(1,length(mtcars$mpg)), y=mtcars$mpg, xlab="Car", ylab="MPG")
abline(h=mean(mtcars$mpg))
``````

This provides our *yardstick* to measure variation from. The overall distance of the data from the mean tells us how much variation there is to explain in this dataset. For instance, the distances for the first and seventh cars are

``````r
mtcars$mpg[1] - mean(mtcars$mpg)
mtcars$mpg[7] - mean(mtcars$mpg)
``````

In order to get an *overall* sense of the variation across all the cars, we might first think to *sum* these distances. However, if we do that then we get a result that is essentially 0

``````r
sum(mtcars$mpg - mean(mtcars$mpg))
``````

The problem is that the mean is, by definition, perfectly in the middle of all the data and thus all the positve and negative distances will cancel and we will be left with nothing. As such, we instead *square* the values[^foot1] before summing them

``````r
sum((mtcars$mpg - mean(mtcars$mpg))^2)
``````

``````r
mean((mtcars$mpg - mean(mtcars$mpg))^2)
var(mtcars$mpg, use=)
``````

Formally, the variance of the *random variable* $x$ is defined as

$$
\text{Var}\left(x\right) = E\left[\left(x - E\left[x\right]\right)^{2}\right]
$$

## Normal Probability Models

We can think of the *mean* as a very basic model of the data. It is a model that predicts a constant value for `mpg` from every car. The degree to which this is not true indicates the degree to which the data is *not* constant and thus quantifies the raw amount of variance in the data.

## Statistical Models

The fundamental elements of parametric statistical model are therefore:

- An assumed probability distribution for the data
- A mean function containing parameters that (potentially) connect the expected value of the distribution to other variables
- A variance function that captures the deviations from the expected value (the errors)

Our basic example of using the mean as a model would result in

$$
y_{i} \sim \mathcal{N}\left(\mu,\sigma\right)
$$
